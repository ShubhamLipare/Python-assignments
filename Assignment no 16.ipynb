```python
#Q1
def cycleSort(array):
   writes = 0

   for cycleStart in range(0, len(array) - 1):
      item = array[cycleStart]
   
      pos = cycleStart
      for i in range(cycleStart + 1, len(array)):
         if array[i] < item:
            pos += 1
    
      if pos == cycleStart:
         continue
    
      while item == array[pos]:
         pos += 1
      array[pos], item = item, array[pos]
      writes += 1
 
      while pos != cycleStart:
  
         pos = cycleStart
         for i in range(cycleStart + 1, len(array)):
            if array[i] < item:
               pos += 1

         while item == array[pos]:
            pos += 1
         array[pos], item = item, array[pos]
         writes += 1
   return writes
arr = [1,5,3,4,8,6,3,4,5]
n = len(arr)
cycleSort(arr)
print("Sorted array is : ")
for i in range(0, n) :
   print(arr[i], end = " ")
```

```python
#Q2
def stoogesort(arr, l, h):
 if l >= h:
  return

 if arr[l]>arr[h]:
  t = arr[l]
  arr[l] = arr[h]
  arr[h] = t

 if h-l+1 > 2:
  t = (int)((h-l+1)/3)

  stoogesort(arr, l, (h-t))
 

  stoogesort(arr, l+t, (h))

  stoogesort(arr, l, (h-t))
 
 

arr = [2, 4, 5, 3, 1]
n = len(arr)
 
stoogesort(arr, 0, n-1)
 
for i in range(0, n):
 print(arr[i], end = ' ')
```

```python
#Q5
def pattern(n):
 
    for i in range(1,n+1):
 
        k =i + 1 if(i % 2 != 0) else i

        for g in range(k,n):
            if g>=k:
                print(end="  ")

        for j in range(0,k):
            if j == k - 1:
                print(" * ")
            else:
                print(" * ", end = " ")
 
n = 10
pattern(n)
```

```python
#Q9
def convert24(str1):
     
    if str1[-2:] == "AM" and str1[:2] == "12":
        return "00" + str1[2:-2]
 
    elif str1[-2:] == "AM":
        return str1[:-2]
     
    elif str1[-2:] == "PM" and str1[:2] == "12":
        return str1[:-2]
         
    else:
        return str(int(str1[:2]) + 12) + str1[2:8]
       
print(convert24("08:05:45 PM"))
```
